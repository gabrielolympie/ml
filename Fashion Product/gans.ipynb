```python
import imageio
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
import cv2
import _pickle as pickle
```

```python
def save(file,name, folder = ""):
    if folder != "":
        outfile = open('./'+folder+'/'+name+'.pickle', 'wb')
    else:
        outfile = open(name+'.pickle', 'wb')
    pickle.dump(file, outfile)
    outfile.close
    
def load(name, folder = ""):
    if folder != "":
        outfile = open('./'+folder+'/'+name+'.pickle', 'rb')
    else:
        outfile = open(name+'.pickle', 'rb')
    file = pickle.load(outfile)
    outfile.close
    return file
```

```python
df = pd.read_csv('styles.csv', delimiter = ';')
def app(x):
    return str(x)+'.jpg'
df['image'] = df['id'].apply(app)

tot_list = os.listdir('./image_clean_128')

df1 = pd.DataFrame({'name':tot_list})
df1['present'] = 1

df.index = df['image'].values
df1.index = df1['name'].values

df = df.join(df1, how = 'left')

df['present'] = df['present'].fillna(0)

df['masterCategory'].value_counts()
```
