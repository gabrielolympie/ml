```python
for fold in tqdm(os.listdir('./raw data')):
    df = []
    for elt in os.listdir('./raw data/'+fold):
        name = elt.split('_')[0] + '.csv'
        df.append(pd.read_csv('./raw data/'+fold + '/'+ elt, sep  ='	', encoding = 'latin-1'))
    df = pd.concat(df)
    df.to_csv('./Agg/'+name, sep = ';')
        
```
