```python
import bs4
import urllib.request as urllib2
from bs4 import BeautifulSoup
```

```python
quote_page = 'https://www.boursier.com/actualites/news-du-jour/20050520'
```

```python
page = urllib2.urlopen(quote_page)
```

```python
soup = BeautifulSoup(page, 'html.parser')
```

```python
soup
```

```python
name_box = soup.find_all('div', attrs={'class': 'item'})
```

```python
len(name_box)
```

```python
a = name_box[27]
```

```python
date = str(a.find('time', attrs={'class': 'date'}))
date = date.replace('<', ' ').replace('>', ' ').split(' ')

for elt in date:
    if 'datetime' in elt:
        tok = elt.split('=')
        print('the ' + tok[0]+ ' is : '+tok[1])
```

```python
b = str(a.find('a', attrs={'class': ''}))
b = b.replace('href', '|href').replace('<b>', '').replace('</b>', '').replace('<', '|').replace('>', '|').split('|')

c = 0

for elt in b:
    if c == 1:
        c = 0
        print('the title of the article is : ' + str(elt))
        
    if 'href' in elt:
        link = elt.split('=')
        print('the ' + link[0]+ ' is : '+ 'https://www.boursier.com'+link[1].replace('"', ''))
        c = 1
        
        follow = 'https://www.boursier.com'+link[1].replace('"', '')
    
    
```

```python
page_child = urllib2.urlopen(follow)
soup_child = BeautifulSoup(page_child, 'html.parser')

article = soup_child.find_all('div', attrs={'class': 'main-article__body'})[0]
article = article.text.split('\n')
M = 0
content = ''
for elt in article:
    if len(elt)>M:
        M = len(elt)
        content = elt.replace("\'", "'")
content
```
