```python
def objective(trial, X_train = X_train, X_test= X_test, y_train=y_train, y_test=y_test):
    reg_param = trial.suggest_uniform('reg',0.01, 0.99)
    classifier_obj = QuadraticDiscriminantAnalysis(reg_param=reg_param)
    clf = classifier_obj.fit(X_train,y_train)
    pred = clf.predict_proba(X_test)[:,1]
    
    return 1-roc_auc_score(y_test, pred)
```

```python
study = optuna.create_study()
```

```python
study.optimize(objective, n_trials=30)
```

```python
clf = QuadraticDiscriminantAnalysis()
clf.reg_param = study.best_params['reg']


clf.fit(X_train, y_train)
pred = clf.predict_proba(X_test)[:,1]
roc_auc_score(y_test, pred)
```
