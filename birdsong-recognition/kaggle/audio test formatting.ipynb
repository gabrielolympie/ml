```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('example_test_audio_metadata.csv')
```

```python
df.head()
```

```python
files_id = df['file_id'].unique()
```

```python
site = []
row_id = []
seconds = []
audio_id = []
labels = []


for file in files_id:
    df1 = df[df['file_id'] == file]
    
    ids = file
    
    max_time = int((df1['time_end'].max()//5 + 1) * 5)
    
    for time_windows in range(5,max_time, 5):
        site.append('site_1')
        seconds.append(time_windows)
        audio_id.append(ids)
        row_id.append('site_1_'+ids+'_'+str(time_windows))
        
        df2 = df1[((df1['time_start']<=time_windows)&(df1['time_start']>time_windows-5))|((df1['time_end']<=time_windows)&(df1['time_end']>time_windows-5))]
        
        birds = df2['ebird_code'].unique()
        labels.append(' '.join(birds))

test = pd.DataFrame({'site':site, 'row_id':row_id, 'seconds': seconds, 'audio_id': audio_id, 'labels': labels})
```

```python
test.head(50)
```

```python
test.to_csv('test_example.csv', index = False)
```
