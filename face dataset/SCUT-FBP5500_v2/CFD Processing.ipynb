```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
from PIL import Image
import imageio
from copy import deepcopy
```

```python
def process_save(im_name, path = './Images'):
    
    import imageio
    import cv2
    import os
    from copy import deepcopy
    import numpy as np
    def resize(img, size):
        img1 = deepcopy(img)
        x_size = img1.shape[0]
        y_size = img1.shape[1]

        if x_size > y_size:
            add = np.zeros((x_size, int((x_size-y_size)/2), 3))+255
            img1 = np.concatenate([add,img1, add], axis = 1)
        else:
            add = np.zeros((int((y_size-x_size)/2), y_size, 3))+255
            img1 = np.concatenate([add,img1, add], axis = 0)
        img1 = cv2.resize(img1, (size,size))
        return img1.astype('uint8')
    
    img = imageio.imread(path+'/'+im_name)
    img1 = resize(img, 256)
    imageio.imsave('./img_256/'+im_name, img1) 
    return im_name
```

```python
from multiprocess import Pool

p = Pool(6)

a = p.map(process_save, os.listdir('./Images'))

p.close
```
