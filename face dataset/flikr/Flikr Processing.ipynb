```python
  
def handle(folder):
    import imageio
    import cv2
    import os
    from copy import deepcopy
    import numpy as np
    def resize(img, size):
        img1 = deepcopy(img)
        x_size = img1.shape[0]
        y_size = img1.shape[1]

        if x_size > y_size:
            add = np.zeros((x_size, int((x_size-y_size)/2), 3))+255
            img1 = np.concatenate([add,img1, add], axis = 1)
        else:
            add = np.zeros((int((y_size-x_size)/2), y_size, 3))+255
            img1 = np.concatenate([add,img1, add], axis = 0)
        img1 = cv2.resize(img1, (size,size))
        return img1.astype('uint8')

    def process_save(im_name, path):
        img = imageio.imread(path+'/'+im_name)
        img1 = resize(img, 256)
        imageio.imsave('./img_256/'+im_name.split('.')[0]+'.jpg', img1)    
    
    
    folder = './raw_images/'+folder
    for elt in os.listdir(folder):
        if elt.split('.')[-1] in ['jpeg', 'jpg', 'png']:
            process_save(elt, folder)
            print(elt)
    return folder
```

```python
from multiprocess import Pool

p = Pool(6)

a = p.map(handle, os.listdir('./raw_images'))

p.close()
```
