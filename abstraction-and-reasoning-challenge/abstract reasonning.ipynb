```python
train = []


task = 0
for file in os.listdir('./training/'):
    with open('./training/'+file, 'r') as myfile:
        data=myfile.read()
    obj = json.loads(data)
    
    training = obj['train']
    testing = obj['test']
    
    task_id = file[:8]
    sub_task = 0
    target = 'train'
    for elt in training:
        inputs = elt['input']
        outputs = elt['output']
        
        vect = [task, task_id, sub_task,target,  inputs, outputs]
        
        train.append(vect)
        
        sub_task += 1
    
    sub_task = 0
    target = 'test'
    for elt in testing:
        inputs = elt['input']
        outputs = elt['output']
        
        vect = [task, task_id, sub_task,target,  inputs, outputs]
        
        train.append(vect)
        
        sub_task += 1        
        
    task += 1

df = pd.DataFrame(train, columns = ['task', 'task_id', 'sub_task', 'target', 'input', 'output'])
```

```python
def plot(task):
    df1 = df[df['task'] == task]
    
    for i in range(df1.shape[0]):
        in1 = df1.iloc[i]['input']
        out1 = df1.iloc[i]['output']
        
        st = df1.iloc[i]['sub_task']
        t = df1.iloc[i]['target']
        
        fig, (ax1, ax2) = plt.subplots(1, 2)
        fig.suptitle(str(t) + ' ' + str(st))
        ax1.imshow(in1)
        ax2.imshow(out1)
```

```python
plot(22)
```
